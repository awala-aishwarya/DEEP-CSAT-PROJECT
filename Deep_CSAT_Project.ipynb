## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation

df.isnull().sum()
````

````python
# Filling missing values

df['Customer Remarks'].fillna("No remarks", inplace=True)

df['Item_price'].fillna(df['Item_price'].median(), inplace=True)

df['connected_handling_time'].fillna(df['connected_handling_time'].mean(), inplace=True)
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

Missing values were handled using different imputation techniques based on the data type.
For numerical variables such as Item_price and connected_handling_time, mean and median imputation were used to preserve the distribution of the data. For the text column Customer Remarks, missing values were replaced with a placeholder text (“No remarks”). This approach ensures that the dataset does not contain null values that may cause errors during model training.

### 2. Handling Outliers

````python
# Handling Outliers using IQR method

Q1 = df['Item_price'].quantile(0.25)
Q3 = df['Item_price'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

df = df[(df['Item_price'] >= lower) & (df['Item_price'] <= upper)]
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Outliers were detected using the Interquartile Range (IQR) method.
The IQR technique identifies extreme values that fall outside the acceptable range of the data distribution. Removing these outliers helps improve model performance and prevents extreme values from distorting statistical analysis.

### 3. Categorical Encoding

````python
# Encode categorical columns

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cat_cols = ['channel_name','category','Sub-category','Customer_City',
            'Product_category','Agent_name','Supervisor','Manager',
            'Tenure Bucket','Agent Shift']

for col in cat_cols:
    df[col] = le.fit_transform(df[col])
````

### 4. Textual Data Preprocessing

#### 1. Expand Contraction

#### 4. Removing URLs & Removing words and digits contain digits.

````python
# Remove URLs & Remove words and digits contain digits
import re

df['Customer Remarks'] = df['Customer Remarks'].apply(lambda x: re.sub(r"http\S+|www\S+", "", x))
df['Customer Remarks'] = df['Customer Remarks'].apply(lambda x: re.sub(r'\d+', '', x))
````

#### 5. Removing Stopwords & Removing White spaces

````python
import nltk

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
````

````python
# Remove Stopwords
from nltk.corpus import stopwords

stop = stopwords.words('english')

df['Customer Remarks'] = df['Customer Remarks'].apply(lambda x: " ".join([w for w in x.split() if w not in stop]))
````

````python
# Remove White spaces
df['Customer Remarks'] = df['Customer Remarks'].str.strip()
````
